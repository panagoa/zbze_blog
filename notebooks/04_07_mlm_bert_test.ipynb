```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("./mlm_bert")
input_text = "Мадинэ и нэпс пIащэ пщтырхэр Iуву и напэм къытелъадэри, зэрыгъыр зыхримыгъэхыну, и IэмкIэ и жьэр иубыдыжащ"
print(tokenizer.tokenize(input_text))

print(len(input_text.split()))
print(len(tokenizer.tokenize(input_text)))
```

```python
from transformers import AutoTokenizer, AutoModel, utils
from bertviz import model_view
utils.logging.set_verbosity_error()  # Suppress standard warnings

model_name = "./mlm_bert"
model = AutoModel.from_pretrained(model_name, output_attentions=True)  # Configure model to return attention values
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
input_text = "Мадинэ и нэпс пIащэ пщтырхэр Iуву и напэм къытелъадэри, зэрыгъыр зыхримыгъэхыну, и IэмкIэ и жьэр иубыдыжащ"  
inputs = tokenizer.encode(input_text, return_tensors='pt')  # Tokenize input text
outputs = model(inputs)  # Run model
attention = outputs[-1]  # Retrieve attention from model outputs
tokens = tokenizer.convert_ids_to_tokens(inputs[0])  # Convert input ids to token strings
model_view(attention, tokens)  # Display model view
```
