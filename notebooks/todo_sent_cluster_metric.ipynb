```python
from collections import defaultdict
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer

sentences = """
Лъагъуныгъэм тIури щызэхэтщ.
А псори жат, хьэдэ нэхъей.
АдэкIэ къэхъуар псоми дощIэ….
– къыгурыIуакIэт абы псори.
А псори щIишэчыжыр сыт?
Дахэщ псори, нэр щыджылу.
Хэт псори зэхэзыщIыхьар?
Псори тэмэм хъунщ, умыгузавэ!
Ауэ сэ псори сщыгъупщэжащ.
Абы псори къикIыркъэ?..
Тэрэзу зэфIэкIащ псори.
Абы псори къыгуроIуэ.
Псори тэмэм хуэдэу къысщохъу.
Псори Сэ щэхуу сщIынщ тэрэз.
Абдежым псори мэдыхьэшх.
Мис псори къыппоплъэ!
Ауэ а псори гум къинэжыркъым.
ЩыIэжкъым зыри, псори пцIыщ.
Абы псори тэмэму зэпилъытат.
А псори хэбдзын хэмылъу пэжт.
Сэ псори къызгуроIуэ.
Хьэуэ, псори хъарзынэщ.
Арат псори зэлъытыжар.
Зэманым псори ехъуэж.
Абы псори къигъэщIащ.
Псори фIыуэ солъагъу сэ!
Ауэ а псори пшынауэ хъуакъым.
ИтIани, а псори зэрегъэуIуф.
Ар есат псори къехъуапсэу.
ЖаIэну псори жаIа хуэдэщ.
Псори екIуэкIащ, зэфIэкIащ.
– Аращ псори абы зэреджэр.
Езым жриIэжат псори Марат.
Апчарэ псори фIэгъэщIэгъуэнт.
Арат псори зэгупсысыр.
Псори сщIэну сыхуейщ.
псори зэхикъутэрти бетэмалу.
Уэ псоми ущыгъуазэкъым.
– Пэжу псори къыбгурыIуащ.
Уэращ псори зыхузэфIэкIынур.
Псори зэтес дыдэ хъужынут.
Алэдамэ къыгурыIуэрт а псори.
""".split('\n')[1:-1]

model = SentenceTransformer('./sbert_from_mlm_bert_6')

embeddings = model.encode(sentences)


clustering_model = KMeans(n_init=10, n_clusters=4, max_iter=1000, random_state=42)
clustering_model.fit(embeddings)
cluster_assignment = clustering_model.labels_


clusters = defaultdict(list)
for sentence, cluster_id in zip(sentences, cluster_assignment):
    clusters[cluster_id].append(sentence)
    
clusters
```
