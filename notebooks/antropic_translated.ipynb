```python
import pandas as pd
import os
```

```python
tr_dir = '../data/processed/translated/sentences/' 

df = pd.DataFrame()
for dir_i in os.listdir(tr_dir):
    for file in os.listdir(os.path.join(tr_dir, dir_i)):
        print(dir_i, file)
        df_i = pd.read_csv(os.path.join(tr_dir, dir_i, file))
        df = pd.concat([df, df_i], axis=0)
```

```python
df.sort_values('source', inplace=True)
```

```python
df = df[df['translated'] != '""']
df = df.drop_duplicates(subset=['source', 'translated'])
```

```python
df.to_csv('../data/processed/translated/sentences/sents_claude-3-5-sonnet-20240620.csv', index=False)
```

```python
word_dir = '../data/processed/word_freqs/'

words_df = pd.DataFrame()

for file in os.listdir(word_dir):
    if 'claude-3-5-sonnet-20240620' not in file:
        continue

    df_i = pd.read_csv(os.path.join(word_dir, file))
    words_df = pd.concat([words_df, df_i], axis=0)
```

```python
words_df.to_csv('../data/processed/word_freqs/words_claude-3-5-sonnet-20240620.csv', index=False)
```
